```python
df1_dummy = pd.get_dummies(df1, drop_first=False)
df1_dummy
```

```python
x = df1_dummy.drop('events',axis=1)                   
y = df1_dummy['events']
```

```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, jaccard_similarity_score, classification_report, log_loss

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=1)
print ('Train set:', x_train.shape,  y_train.shape)
print ('Test set:', x_test.shape,  y_test.shape)
```

```python
opt = ['newton-cg','lbfgs','liblinear','sag','saga']

for i in opt:
    LR = LogisticRegression(C=1, solver=i).fit(x_train,y_train)
    yhat = LR.predict(x_test)
    yhat_proba = LR.predict_proba(x_test)
    print("solver = ", i)
    print(" ")
    print("Jaccard = ", jaccard_similarity_score(y_test, yhat))
    print(" ")
    print(classification_report(y_test, yhat))
    print(" ")
    print("Log Loss = ", log_loss(y_test, yhat_proba))
    print(" ")
    print("--------------------------------------------------------")
    print(" ")
```

```python
LR_best = LogisticRegression(C=1, solver='newton-cg').fit(x_train,y_train)
yhat_best = LR_best.predict(x_test)
```

```python
import itertools
def plot_confusion_matrix(cm, classes,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
print(confusion_matrix(y_test, yhat_best, labels=[1,0]))
```

```python
cnf_matrix = confusion_matrix(y_test, yhat_best, labels=[1,0])
np.set_printoptions(precision=2)

plt.figure()
plot_confusion_matrix(cnf_matrix, classes=['event=1','event=0'],  title='Confusion matrix')
```

```python
from sklearn.model_selection import cross_val_score
rcross = cross_val_score(LR_best, x, y, cv=4)
print("R^2 =",rcross.mean())
```

# 3. Monthly rate of discontinuation

```python
rod = df.loc[(df['Prov'].isin(['ALL'])) & (df['Con_ACT'].isin(['ALL'])) & (df['Sex'].isin(['ALL'])) & (df['Age'].isin(['ALL']))]
rod.reset_index(drop=True, inplace=True)
```

```python
columns=range(len(rod.columns)-5)
```

```python
dis = 0

for i in columns:
    if i <= 38:
        x = i+6
        dis += rod.iloc[2,x]/rod.iloc[0,x-1]
        
    else:
        avg_dis_pct = (dis/39)*100
        print("The monthly rate of discontinuation is", round(avg_dis_pct,2),"%")
    
```
